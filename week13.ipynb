```python
import matplotlib.pyplot as plt
import csv
f= open('seoul2.csv')
data = csv.reader(f)
header = next(data)
high = []
low = []
for row in data :
    if len(row)>1:
        if row[-1]!='' and row[-2]!='':
            if row[0].split('-')[1] == '02' and row[0].split('-')[2] == '28':
                high.append(float(row[-1]))
                low.append(float(row[-2]))
plt.plot(high,'hotpink')
plt.plot(low, 'skyblue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
f = open('seoul2.csv')
data = csv.reader(f)
header = next(data)
aug = []
jan = []

for row in data :
    if len(row)>1:
        month = row[0].split('-')[1]
        if row[-1] != '':
            if month == '08':
                aug.append(float(row[-1]))
            elif month == '01':
                jan.append(float(row[-1]))
plt.hist(aug,bins=100,color='r',label='Aug')
plt.hist(jan,bins=100,color='blue',label='Jan')
plt.legend()
plt.show()
```
