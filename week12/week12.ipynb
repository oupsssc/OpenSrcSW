```python
#실습2
import csv

def h_sort(li):
    for i in reversed(range(len(li)-1)):# 2개들어있을때 0, 3개들어있을 때 1,0
        #print(li)
        tmp1 = li[i+1]
        tmp2 = li[i]
        if(tmp1[1]>tmp2[1]):
            #print(tmp1[1])
            #print(tmp2[1])
            tmp3 = tmp1
            li[i+1] = tmp2
            li[i] = tmp3
        #print(li)
        #print("")

            
def l_sort(li):
    for i in reversed(range(len(li)-1)):# 2개들어있을때 0, 3개들어있을 때 1,0
        tmp1 = li[i+1]
        tmp2 = li[i]
        if(tmp1[1]<tmp2[1]):
            tmp3 = tmp1
            li[i+1] = tmp2
            li[i] = tmp3



lines_high = {}#노선 : 3개를 가지고 있는 리스트
lines_low = {} 
f= open('subwayfee.csv','r',encoding='cp949')
data = csv.reader(f)
header= next(data)

for row in data :
    row[-2] = row[-3].replace(",","")
    row[-3] = row[-3].replace(",","")
    if row[-2]=='':
        row[-2]=0
    if row[-3]=='':
        row[-3]=0
    row[-2] = int(row[-2])
    row[-3] = int(row[-3])
    num = row[-2] + row[-3]
    t = (row[3],num)
    if row[1] in lines_high:
        if len(lines_high[row[1]])<4:
            li = lines_high[row[1]]
            lines_high[row[1]].append(t)
            h_sort(li)
        else:#4개가 차있을때
            li = lines_high[row[1]]
            li[3] = t
            h_sort(li)
    else:
        lines_high[row[1]] = [t] 
        
    if row[1] in lines_low:
        if len(lines_low[row[1]])<4:
            li = lines_low[row[1]]
            lines_low[row[1]].append(t)
            l_sort(li)
        else:#4개가 차있을때
            li = lines_low[row[1]]
            li[3] = t
            l_sort(li)
    else:
        lines_low[row[1]] = [t] 
        
f.close()
for k in lines_high.keys():
    print("<",k,">")
    print("이용객이 가장 많은 역 순위")
    li_high = lines_high[k]
    li_low = lines_low[k]
    for i in range(3):
        tmp = li_high[i]
        print(i+1,".",tmp[0],"(",tmp[1],")")
    print("이용객이 가장 적은 역 순위")
    for i in range(3):
        tmp = li_low[i]
        print(i+1,".",tmp[0],"(",tmp[1],")")
    print("")



```
